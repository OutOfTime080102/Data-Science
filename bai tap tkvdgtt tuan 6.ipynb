4. Bài toán người bán hàng.

```python

from sys import maxsize
from itertools import permutations
V = 4

# implementation of traveling Salesman Problem
def travellingSalesmanProblem(graph, s):

    # store all vertex apart from source vertex
    vertex = []
    for i in range(V):
        if i != s:
            vertex.append(i)

    # store minimum weight Hamiltonian Cycle
    min_path = maxsize
    next_permutation=permutations(vertex)
    for i in next_permutation:

        # store current Path weight(cost)
        current_pathweight = 0

        # compute current path weight
        k = s
        for j in i:
            current_pathweight += graph[k][j]
            k = j
        current_pathweight += graph[k][s]

        # update minimum
        min_path = min(min_path, current_pathweight)

    return min_path


# Driver Code
if __name__ == "__main__":

    # matrix representation of graph
    graph = [[0, 10, 15, 20], [10, 0, 35, 25],
            [15, 35, 0, 30], [20, 25, 30, 0]]
    s = 0
    print(travellingSalesmanProblem(graph, s))
```

1. Khớp xâu: Tìm các vị trí xuất hiện của xâu mẫu P trong văn bản T cho
trước

```python
def find(P, T):
    m, n = len(P), len(T)
    lps = [0] * m
    computeLPS(P, m, lps)
    i, j, res = 0, 0, []
    while i < n:
        if P[j] == T[i]:
            i, j = i + 1, j + 1
        if j == m:
            res.append(i - j)
            j = lps[j - 1]
        elif i < n and P[j] != T[i]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1
    return res

def computeLPS(P, m, lps):
    len = 0
    lps[0] = 0
    i = 1
    while i < m:
        if P[i] == P[len]:
            len += 1
            lps[i] = len
            i += 1
        else:
            if len != 0:
                len = lps[len - 1]
            else:
                lps[i] = 0
                i += 1
# ví dụ :
P = "ys"
T = "auminysuiosybnhysiopbnysuionm"
matches = find(P, T)
print(f"Các vị trí xuất hiện của xâu mẫu '{P}' trong văn bản '{T}' là:")
print(matches)
```

Liệt kê các hoán vị của n phần tử.

```python
def permutations(n):
    nums = list(range(1, n + 1))
    result = []

    def backtrack(first):
        if first == n:
            result.append(nums[:])
        else:
            for i in range(first, n):
                nums[first], nums[i] = nums[i], nums[first]
                backtrack(first + 1)
                nums[first], nums[i] = nums[i], nums[first]

    backtrack(0)
    return result

# Ví dụ n = 6
n = 6
perms = permutations(n)
print(f"Tất cả các hoán vị của {n} phần tử là:")
for perm in perms:
    print(perm)
```

Liệt kê các tổ hợp k phần tử của các số từ 1 đến n.

```python
def combine(n, k):
    result = []

    def backtrack(first, combo):
        if len(combo) == k:
            result.append(combo[:])
        else:
            for i in range(first, n + 1):
                combo.append(i)
                backtrack(i + 1, combo)
                combo.pop()

    backtrack(1, [])
    return result

# Ví dụ n = 6, k =2
n = 6
k = 2
combos = combine(n, k)
print(f"Tất cả các tổ hợp {k} phần tử của các số từ 1 đến {n} là:")
for combo in combos:
    print(combo)
```
